## **Proof of Central Limit Theorem**

**Importing Libraries:** The code begins by importing the necessary libraries. numpy is imported as np for numerical operations and random number generation, and matplotlib.pyplot is imported as plt for data visualization.

**Simulation Parameters:** The parameters for the simulation are defined at the beginning. num_simulations determines the number of times the entire simulation will be run, num_samples specifies the number of samples taken in each simulation, and sample_size defines the size of each individual sample.

**Generating Normal Data:** The generate_normal_data function is defined to simulate data from a normal distribution. It takes the parameter n, which represents the number of data points to generate. The function uses np.random.normal to generate random data points with a mean of 0 and a standard deviation of 1, adhering to the standard normal distribution.

**Simulating Sample Means:** The outer loop runs the simulation num_simulations times. Within each simulation, an inner loop generates num_samples samples of size sample_size each from the normal distribution using the generate_normal_data function. For each sample, the mean is calculated and stored in the sample_means list.

**Aggregating Sample Means:** The calculated sample means from all simulations are aggregated into the all_sample_means list. This list will hold the sample means from all simulations combined.

**Plotting the Distribution:** Using matplotlib, a histogram of the all_sample_means is plotted. The plt.hist function is used to create a histogram, where bins specifies the number of bins or intervals in the histogram, edgecolor sets the color of the edges of the bars, and density=True normalizes the histogram to represent a probability density.

**Plot Customization:** Titles, labels, and layout adjustments are added to the plot using plt.title, plt.xlabel, plt.ylabel, and plt.tight_layout.

**Displaying the Plot:** Finally, the plot is displayed using plt.show().

The code demonstrates the Central Limit Theorem by simulating the distribution of sample means from a normal distribution. As the number of samples and simulations increases, the resulting histogram of sample means approaches a bell-shaped curve, showcasing the Central Limit Theorem's concept.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the simulation
num_simulations = 10000   # Number of simulations
num_samples = 1000       # Number of samples in each simulation
sample_size = 20         # Size of each sample

# Simulate data from a normal distribution
def generate_normal_data(n):
    return np.random.normal(0, 1, n)

# Simulate the distribution of the sample means for multiple simulations
all_sample_means = []
for _ in range(num_simulations):
    sample_means = []
    for _ in range(num_samples):
        sample = generate_normal_data(sample_size)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)
    all_sample_means.extend(sample_means)

# Plot the distribution of all sample means
plt.figure(figsize=(10, 6))

plt.hist(all_sample_means, bins=100, edgecolor='black', density=True)
plt.title('Distribution of Sample Means for Multiple Simulations (Normal Distribution)')
plt.xlabel('Mean Value')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```
